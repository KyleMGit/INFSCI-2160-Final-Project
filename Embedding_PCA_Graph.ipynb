```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import plotly.express as px
```

```python
df = pd.read_pickle('df_with_embeddings.pkl')
embeddings = np.stack(df['embeddings'].values)
df['simple_bias'] = 'center'
df.loc[(df['bias'] == 'leaning-left') | (df['bias'] == 'left'), 'simple_bias'] = 'left'
df.loc[(df['bias'] == 'leaning-right') | (df['bias'] == 'right'), 'simple_bias'] = 'right'
# 3D PCA
pca = PCA(n_components=3)
components = pca.fit_transform(embeddings)

print(f"Explained variance: {pca.explained_variance_ratio_.sum():.2%}")

pca_df = pd.DataFrame({
    'PC1': components[:, 0],
    'PC2': components[:, 1],
    'PC3': components[:, 2],
    'bias': df['simple_bias']
})

fig = px.scatter_3d(
    pca_df,
    x='PC1', y='PC2', z='PC3',
    color='bias',
    color_discrete_map={'left': '#4C72B0', 'center': '#8C8C8C', 'right': '#C44E52'},
    opacity=0.6,
    title='PCA of News Article Embeddings',
    labels={'PC1': f'PC1 ({pca.explained_variance_ratio_[0]:.1%})',
            'PC2': f'PC2 ({pca.explained_variance_ratio_[1]:.1%})',
            'PC3': f'PC3 ({pca.explained_variance_ratio_[2]:.1%})'}
)

fig.update_traces(marker=dict(size=3))
fig.update_layout(legend_title_text='Bias')
fig.show()
```

```python
fig.write_html("pca_bias_visualization.html")
```
